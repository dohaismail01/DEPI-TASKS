insights for each visualization 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
# 1. Survival count
plt.figure()
titanic['survived'].value_counts().sort_index().plot(kind='bar')
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

Survival Count Bar Chart
Shows how many passengers survived vs died.

Insight:
Usually more people died than survived → highlights severity of disaster.

```python
# 2. Age distribution
plt.figure()
titanic['age'].dropna().plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Age Distribution Histogram
Shows how passenger ages are distributed.

Insight:
Most passengers are young adults (20–40)
Fewer elderly and very young passengers

```python

# 3. Survival rate by class
plt.figure()
titanic.groupby('class')['survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Class")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.show()
```

Survival Rate by Class
Shows average survival per class.

Insight:
1st class -> highest survival
3rd class ->lowest survival
Indicates social/economic status influenced rescue priority.

```python
# 4. Fare vs Age
plt.figure()
plt.scatter(titanic['age'], titanic['fare'])
plt.title("Fare vs Age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

Fare vs Age Scatter Plot
Shows relationship between ticket price and age.

Insight:
Wide fare range for adults
Expensive tickets often linked to higher class passengers
No strong age fare correlation, but class matters.

```python
plt.figure()
sns.heatmap(titanic.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

insight:
fare > positive correlation
pclass > negative correlation (higher class number = lower survival)
age > slight negative
sibsp, parch > small effect

```python
plt.figure()
sns.boxplot(x='class', y='age', data=titanic)
plt.title("Age Distribution across Passenger Classes")
plt.show()
```

1st class passengers tend to be older

3rd class has younger passengers and children

Indicates wealthier passengers were older.
